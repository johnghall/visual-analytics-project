```python
# IMPORTS
import pandas as pd
import numpy as np
from vega_datasets import data
import altair as alt
```

```python
# CENSUS DATA CLEAN
census_data = pd.read_csv('./census.csv')
census_data = census_data[['FID', 'NAME', 'B01001_calc_PopDensity', 'State', 'GEOID']]
census_data = census_data.rename(columns={'B01001_calc_PopDensity': 'DENSITY(N)', 'GEOID': 'id'})

def label_density(row):
  if row['DENSITY(N)'] >= 200:
    return 'Urban'
  elif row['DENSITY(N)'] >= 100:
    return 'Suburban'
  else:
    return 'Rural'

census_data['DENSITY(C)'] = census_data.apply( lambda row: label_density(row), axis=1 )
census_data.to_csv("population_density_data.csv")
```

```python
# CHLOROPLETH FOR COUNTY DENSITY
selector = alt.selection_single()
counties = alt.topo_feature(data.us_10m.url, 'counties')
ch_map = alt.Chart(counties).mark_geoshape(stroke='white').encode(
    color=alt.condition(selector, 'DENSITY(C):N', alt.value('lightgray')),
    tooltip=['NAME:N', 'DENSITY(N):Q', 'DENSITY(C):N', 'State:N']
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(census_data, 'id', ['NAME','DENSITY(N)','DENSITY(C)','State'])
).project(
    type='albersUsa'
).properties(
    width=600,
    height=400
).add_selection(
    selector
)

ch_map
```
